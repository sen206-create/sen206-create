```python
import mne

sample_data_folder = mne.datasets.sample.data_path()
print(sample_data_folder)
```

```python
raw_path = sample_data_folder / "MEG" / "sample" / "sample_audvis_raw.fif"

raw = mne.io.read_raw_fif(raw_path, preload=True)

print(raw)
```

```python
raw.ch_names
```

```python
raw_eeg = raw.copy().pick("eeg")

print(raw_eeg)
print(raw_eeg.ch_names)
```

```python
raw_eeg.plot()
```

```python
raw_eeg.plot(duration=10, n_channels=10)
```

```python
raw_eeg.info["sfreq"]
```

```python
raw_eeg.times[-1]
```

```python
raw_eeg_filtered = raw_eeg.copy().filter(l_freq=1, h_freq=40)
```

```python
raw_eeg_filtered.plot(duration=10, n_channels=10)
```

```python
raw_eeg_filtered.notch_filter(freqs=50)
```

```python
raw_eeg_filtered.info["bads"]
```

```python
raw_eeg_clean = raw_eeg_filtered.copy().interpolate_bads()
```

```python
raw_eeg_clean.info["bads"]
```

```python
raw_eeg_clean.set_eeg_reference("average")
```

```python
raw_eeg_clean.save("cleaned_eeg_raw.fif", overwrite=True)
```

```python
events = mne.find_events(raw)

print(events[:10])
print(events.shape)
```

```python
set(events[:, 2])
```

```python
event_id = {
    "auditory/left": 1,
    "auditory/right": 2,
    "visual/left": 3,
    "visual/right": 4,
    "smiley": 5,
    "button": 32,
}
```

```python
epochs = mne.Epochs(
    raw_eeg_clean,
    events,
    event_id=event_id,
    tmin=-0.2,
    tmax=0.5,
    baseline=(None, 0),
    preload=True
)

print(epochs)
```

```python
evoked_auditory_left = epochs["auditory/left"].average()

evoked_auditory_left.plot()
```

```python
evoked_auditory_right = epochs["auditory/right"].average()

mne.viz.plot_compare_evokeds(
    {
        "auditory/left": evoked_auditory_left,
        "auditory/right": evoked_auditory_right,
    },
    picks="eeg"
)
```

```python
evoked_auditory_left.plot_topomap(times=[0.05, 0.1, 0.15, 0.2])
```

```python
evoked_visual_left = epochs["visual/left"].average()
evoked_visual_right = epochs["visual/right"].average()

mne.viz.plot_compare_evokeds(
    {
        "auditory/left": evoked_auditory_left,
        "visual/left": evoked_visual_left,
    },
    picks="eeg"
)
```

```python
evoked_visual_left.plot_topomap(times=[0.05, 0.1, 0.15, 0.2, 0.25, 0.3])
evoked_auditory_left.plot_topomap(times=[0.05, 0.1, 0.15, 0.2, 0.25, 0.3])
```

```python
auditory_gfp = evoked_auditory_left.data.std(axis=0)
visual_gfp = evoked_visual_left.data.std(axis=0)

auditory_peak_time = evoked_auditory_left.times[auditory_gfp.argmax()]
visual_peak_time = evoked_visual_left.times[visual_gfp.argmax()]

print("Auditory peak time:", auditory_peak_time)
print("Visual peak time:", visual_peak_time)
```

```python
auditory_peak_strength = auditory_gfp.max()
visual_peak_strength = visual_gfp.max()

print("Auditory peak strength:", auditory_peak_strength)
print("Visual peak strength:", visual_peak_strength)
```

```python
print("Auditory peak time:", round(auditory_peak_time * 1000, 2), "ms")
print("Visual peak time:", round(visual_peak_time * 1000, 2), "ms")

print("Auditory peak strength:", round(auditory_peak_strength * 1e6, 2), "µV")
print("Visual peak strength:", round(visual_peak_strength * 1e6, 2),"µV")
```

```python
time_difference = visual_peak_time - auditory_peak_time
strength_difference = auditory_peak_strength - visual_peak_strength

print("Visual peak is later by:", round(time_difference * 1000, 2), "ms")
print("Auditory peak is stronger by:", round(strength_difference * 1e6, 2), "µV")
```

```python
import matplotlib.pyplot as plt

conditions = ["Auditory", "Visual"]
peak_times_ms = [auditory_peak_time * 1000, visual_peak_time * 1000]

plt.bar(conditions, peak_times_ms)
plt.ylabel("Peak time (ms)")
plt.title("Peak response time: auditory vs visual")
plt.show()
```
